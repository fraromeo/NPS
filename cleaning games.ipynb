```python
import pandas as pd
import numpy as np
from tqdm import tqdm
tqdm.pandas()
```

```python
def compute_best_players(x):

  x = x.rstrip(')')
  x = x.split('])])]),')
  x[-1] = x[-1].rstrip(')]')
  x[-1] = x[-1] + ')'

  vec = np.zeros(shape=len(x))

  for index, row in enumerate(x):
    row = row.strip('[')
    row = row + '])])])'

    try:
      dct = dict(eval(row))
    
    except:
      dct = {'result': [ {'@numvotes': 0}, {'@numvotes': 0}]}

    for i in range(len(dct['result'])):
      dct['result'][i] = dict(eval(str(dct['result'][i])))

    dct['result'] = 2 * int(dct['result'][0]['@numvotes']) + int(dct['result'][1]['@numvotes'])

    vec[index] = dct['result']

  max = np.argmax(vec) + 1 if np.max(vec) > 0 else -1

  return max
```

```python
games = pd.read_csv('/content/drive/Othercomputers/Il mio laptop/NPS/games_da_ripulire.csv')
```

```python
games['suggested_num_players_coef'] = games['suggested_num_players'].progress_apply(lambda x: compute_best_players(x))
```
